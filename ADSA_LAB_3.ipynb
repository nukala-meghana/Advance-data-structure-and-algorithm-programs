```python
# Two stacks:
# main_stack ‚Üí stores all performed actions
# redo_stack ‚Üí stores undone actions (so we can redo them later)
main_stack = []
redo_stack = []


# Function to add a new action
def push():
    # Take action from user
    action = input("Enter action to push: ")

    # Add the action to main_stack
    main_stack.append(action)

    # Clear redo stack because new action invalidates old redo history
    redo_stack.clear()

    print(f"‚úÖ Added: {action}")


# Function to undo the last action
def undo():
    # Check if there is anything to undo
    if not main_stack:
        print("‚ö†Ô∏è Nothing to undo.")
        return

    # Remove the last action from main_stack
    action = main_stack.pop()

    # Move it to redo_stack (so it can be redone later)
    redo_stack.append(action)

    print(f"‚Ü©Ô∏è Undo: {action}")


# Function to redo the last undone action
def redo():
    # Check if there is anything to redo
    if not redo_stack:
        print("‚ö†Ô∏è Nothing to redo.")
        return

    # Take last undone action from redo_stack
    action = redo_stack.pop()

    # Add it back to main_stack
    main_stack.append(action)

    print(f"üîÅ Redo: {action}")


# Function to display both stacks
def display():
    print("\nüìã Current State:")
    print("Main Stack (Done actions):", main_stack)
    print("Redo Stack (Undone actions):", redo_stack)
    print("-" * 40)


# ------------------- MAIN MENU LOOP -------------------

while True:
    # Display menu options
    print("\n=== üß† Undo-Redo System ===")
    print("1. Push (Add Action)")
    print("2. Undo Last Action")
    print("3. Redo Last Action")
    print("4. Display Stacks")
    print("5. Exit Program")

    # Get user's choice
    choice = input("Enter your choice (1-5): ")

    # Match choice with operation
    if choice == '1':
        push()
    elif choice == '2':
        undo()
    elif choice == '3':
        redo()
    elif choice == '4':
        display()
    elif choice == '5':
        print("üëã Exiting program... Goodbye!")
        break
    else:
        print("‚ùå Invalid choice! Please enter a number between 1 and 5.")
s
```
